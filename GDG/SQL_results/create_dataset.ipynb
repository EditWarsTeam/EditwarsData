```python
import pandas as pd
import json
from tqdm import tqdm

series_list = []
with open('january.json', 'r') as file:
    for line in tqdm(file):
        d = json.loads(line)
        series = pd.Series({
        "fetchdate_orig": d["fetchdate_orig"],
        "fetchdate_check": d["fetchdate_check"],
        "status": d["status"],
        "page_title": d["page_title"],
        "page_domain_root": d["page_domain_root"],
        "page_url": d["page_url"],
        "changes": d["changes"]
    })
        series_list.append(series)

with open('february.json', 'r') as file:
    for line in tqdm(file):
        d = json.loads(line)
        series = pd.Series({
        "fetchdate_orig": d["fetchdate_orig"],
        "fetchdate_check": d["fetchdate_check"],
        "status": d["status"],
        "page_title": d["page_title"],
        "page_domain_root": d["page_domain_root"],
        "page_url": d["page_url"],
        "changes": d["changes"]
    })
        series_list.append(series)

with open('march.json', 'r') as file:
    for line in tqdm(file):
        d = json.loads(line)
        series = pd.Series({
        "fetchdate_orig": d["fetchdate_orig"],
        "fetchdate_check": d["fetchdate_check"],
        "status": d["status"],
        "page_title": d["page_title"],
        "page_domain_root": d["page_domain_root"],
        "page_url": d["page_url"],
        "changes": d["changes"]
    })
        series_list.append(series)

with open('april.json', 'r') as file:
    for line in tqdm(file):
        d = json.loads(line)
        series = pd.Series({
        "fetchdate_orig": d["fetchdate_orig"],
        "fetchdate_check": d["fetchdate_check"],
        "status": d["status"],
        "page_title": d["page_title"],
        "page_domain_root": d["page_domain_root"],
        "page_url": d["page_url"],
        "changes": d["changes"]
    })
        series_list.append(series)
```

```python
print(len(series_list))
```

```python
df = pd.DataFrame(series_list, columns=["fetchdate_orig", "fetchdate_check", "status", "page_title", "page_domain_root", "page_url", "changes"])
```

```python
import pandas as pd
clean_df = pd.read_json('united_data.json')
```

```python
from datetime import datetime 
def change_time(dt):
  dt_new = dt.replace(' UTC', '')
  return datetime.strptime(dt_new, '%Y-%m-%d %H:%M:%S')

clean_df['fetchdate_orig'] = clean_df['fetchdate_orig'].apply(change_time)
clean_df['fetchdate_check'] = clean_df['fetchdate_check'].apply(change_time)

clean_df
```

```python
media = pd.read_csv("Media_types.csv")
media
```

```python
df_result = pd.merge(clean_df, media, how='left', on='page_domain_root')
df_result
```
